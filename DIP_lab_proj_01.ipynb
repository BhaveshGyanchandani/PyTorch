```python
import cv2
import numpy as np
import matplotlib.pyplot as plt  # Importing plt

# Define image path
image_path = r'D:\web dev backup\Pytorch\Assign1\Input Images1\2.png'

# Load image
gray = cv2.imread(image_path, 0)

# Check if image is loaded correctly
if gray is None:
    print("Error: Could not load image. Check the file path.")
else:
    # Print dtype and shape
    print("Data Type:", gray.dtype)  # Example Output: uint8
    print("Shape:", gray.shape)      # Example Output: (Height, Width)

    # Show original image
    plt.imshow(gray, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    plt.show()

    # Resize the image
    small = cv2.resize(gray, (0, 0), fx=0.5, fy=0.5)

    # Convert to float
    small = small.astype('float32')

    # Apply intensity shift
    intensity_shift = 200
    small += intensity_shift

    # Clip to valid range
    small = np.clip(small, 0, 255)  # FIXED: Using 'small' instead of 'gray'

    # Convert back to uint8
    small = small.astype('uint8')

    # Display resized image
    plt.imshow(small, cmap='gray')
    plt.title('Resized & Shifted Image')
    plt.axis('off')
    plt.show()

```

```python
import cv2
import matplotlib.pyplot as plt

# Image paths
image_paths = [
    r'D:\web dev backup\Pytorch\Assign1\Input Images1\1.jpg',
    r'D:\web dev backup\Pytorch\Assign1\Input Images1\3.jpg',
    r'D:\web dev backup\Pytorch\Assign1\Input Images1\5.png'
]

images = []
negative_images = []

for path in image_paths:
    img = cv2.imread(path, 0)  # Read in grayscale
    if img is None:
        print(f"Error: Could not load image {path}")
        continue  # Skip this image
    
    images.append(img)
    negative_images.append(255 - img)  # Compute negative image

# Check if we have images to process
if not images:
    print("No valid images loaded. Check file paths.")
else:
    # Titles for the images
    titles = []
    for i in range(len(images)):
        titles.append(f'Input Image {i+1}')
        titles.append(f'Negative Image {i+1}')
    
    # Combine images for display
    all_images = [img for pair in zip(images, negative_images) for img in pair]

    # Plot images
    plt.figure(figsize=(12, 8))
    for i in range(len(all_images)):
        plt.subplot(2, len(images), i + 1)
        plt.imshow(all_images[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')

    plt.tight_layout()
    plt.show()

```
