```python
import json
from transformers import AutoTokenizer
import tiktoken

tokenizer = tiktoken.get_encoding("gpt2")  # Change model if needed

def get_avg_token_length(json_path):
    with open(json_path, "r") as f:
        data = json.load(f)
    total_tokens = sum(len(tokenizer.encode(obj["instruction"] + obj["input"] + obj["output"])) for obj in data)
    return total_tokens / len(data)

print("Avg Tokens (1100 samples):", get_avg_token_length("instruction-data.json"))
print("Avg Tokens (200 samples):", get_avg_token_length("Alpaca_data/alpaca_data_52K_1.json"))
```

```python
import json
import random
import os

# Load the original JSON file
input_file = "alpaca_data_52K.json"  # Change this to your actual JSON file
output_folder = "Alpaca_data"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Read JSON data
with open(input_file, "r", encoding="utf-8") as f:
    data = json.load(f)

# Shuffle the data randomly
random.shuffle(data)

# Split data into chunks of 1,000 objects each
chunk_size = 50
total_chunks = len(data) // chunk_size

for i in range(total_chunks):
    chunk = data[i * chunk_size: (i + 1) * chunk_size]  # Get a slice of 1000 objects

    # Define output file name
    output_file = os.path.join(output_folder, f"alpaca_data_52K_{i+1}.json")

    # Save the chunk as a JSON file
    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(chunk, f, indent=4)

    print(f"Saved {output_file} with {len(chunk)} objects.")

print("Splitting complete!  All files are saved in the 'Alpaca_data' folder.")

```
